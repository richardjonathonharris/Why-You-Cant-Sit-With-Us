```python
import networkx as nx
import os
import csv
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (30.0, 20.0)
```

```python
node_list = []

with open('../data/nodes.csv', 'r') as openfile:
    reader = csv.reader(openfile)
    next(reader, None)
    for row in reader:
        node_list.append((row[0], {'name' : row[1]}))
        
print node_list
```

```python
edge_list = []

with open('../data/edges.csv', 'r') as openfile:
    reader = csv.reader(openfile)
    next(reader, None)
    for row in reader:
        if row[1] == '':
            edge_list.append((row[0], row[0]))
        else:
            edge_list.append((row[0], row[1]))
        
print edge_list
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(node_list)
```
